```python
import pandas as pd
import matplotlib.pyplot as plt


Data = pd.read_csv('netflix_titles.csv')

# Data Cleaning
Data.dropna(subset=['type','country', 'release_year', 'rating', 'duration'])

Type = Data['type'].value_counts()

plt.bar(Type.index, Type.values, color=['skyblue', 'orange'])
plt.title('Tv Shows Vs Movie On Netflix')
plt.xlabel('Type')
plt.ylabel('Count')
plt.tight_layout()

plt.show()

Rating = Data['rating'].value_counts()
plt.pie(Rating, labels= Rating.index, autopct='%1.1f%%', startangle=90)
plt.title('Percentage Of Rating')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the dataset
Data = pd.read_csv('netflix_titles.csv')

# Data Cleaning: drop rows with missing values in these columns
Data = Data.dropna(subset=['type', 'country', 'release_year', 'rating', 'duration'])

# Filter only movies
Movie_DF = Data[Data['type'] == 'Movie'].copy()

# Clean and convert duration to integer
Movie_DF['duration'] = Movie_DF['duration'].str.replace(' min', '').astype(int)

# Display the cleaned Movie DataFrame
# print(Movie_DF.head())

plt.hist(Movie_DF['duration'], bins=30, color='black', edgecolor='white')
plt.title('Distrubutation Of Movie Duration')
plt.xlabel('Duration (Minutes)')
plt.ylabel('Number Of Movie')
plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the dataset
Data = pd.read_csv('netflix_titles.csv')

# Data Cleaning
Data = Data.dropna(subset=['type', 'country', 'release_year', 'rating', 'duration'])

# Bar chart for top 10 countries by number of shows
Country_Count = Data['country'].value_counts().head(10)

plt.figure(figsize=(10, 6))
plt.barh(Country_Count.index, Country_Count.values, color='teal')
plt.title('Top 10 Countries By Number Of Shows')
plt.xlabel('Number Of Shows')
plt.ylabel('Country')
plt.tight_layout()
plt.show()

# Group content by year and type
Content_By_Year = Data.groupby(['release_year', 'type']).size().unstack().fillna(0)

# Create 3 subplots
fig, ax = plt.subplots(1, 3, figsize=(18, 5))

# Plot movies
ax[0].plot(Content_By_Year.index, Content_By_Year['Movie'], color='blue')
ax[0].set_title('Movies by Year')
ax[0].set_xlabel('Year')
ax[0].set_ylabel('Count')

# Plot TV shows
ax[1].plot(Content_By_Year.index, Content_By_Year['TV Show'], color='green')
ax[1].set_title('TV Shows by Year')
ax[1].set_xlabel('Year')
ax[1].set_ylabel('Count')

# Combined plot
ax[2].plot(Content_By_Year.index, Content_By_Year['Movie'], color='blue', label='Movies')
ax[2].plot(Content_By_Year.index, Content_By_Year['TV Show'], color='green', label='TV Shows')
ax[2].set_title('Movies vs TV Shows by Year')
ax[2].set_xlabel('Year')
ax[2].set_ylabel('Count')
ax[2].legend()

plt.tight_layout()
plt.show()
```
